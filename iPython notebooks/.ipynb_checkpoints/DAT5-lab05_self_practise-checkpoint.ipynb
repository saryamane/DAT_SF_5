## Linear Regression

Load a subset of the housing data

```python
data = pd.read_csv('data/houses.csv')
```

```python
data.head()
```

```python
data_n = pd.read_csv('data/houses_norm.csv')
```

The dataset contains 4 columns. Our objective is to build a model that is able to predict the price of a house, using the information contained in the other columns (sqft, bedrooms and age)

```python
% pylab inline
```

```python
from pandas.tools.plotting import scatter_matrix
scat = scatter_matrix(data, figsize = (10,10))
```

Run 1. x = sqft, data is raw (so graph units are nice right)

Let''s start with a simple model that uses the house surface to estimate the price

```python
x = data['sqft'].values # Gives you a numpy array
y = data['price'].values
```

```python
X = sm.add_constant(x, prepend=True) # Putting one infront of the vector.
results = sm.OLS(y, X).fit() # Ordinary Least squares, y is the variable we want to predict.
intercept, slope = results.params

r2 = results.rsquared

plt.plot(x,y,'bo')
plt.title("Price vs sqft")
#x1=np.array([min(x), max(x)])
#y1=intercept + slope * x1
#plt.plot(x1, y1, 'r-')
#plt.show



plt.plot(np.array([min(x), max(x)]), intercept + slope * x1,'r-')
plt.show
```

```python
results.summary()
```

Run 2:  x = sqft, data is normalized (so summary stats are nice)

```python
data_n.head()
```

```python
x = data_n['sqft'].values # Gives you a numpy array
y = data_n['price'].values

X = sm.add_constant(x, prepend=True) # Putting one infront of the vector.
results = sm.OLS(y, X).fit() # Ordinary Least squares, y is the variable we want to predict.
intercept, slope = results.params

r2 = results.rsquared

plt.plot(x,y,'bo')
plt.title("Price vs sqft")
x1=np.array([min(x), max(x)])

plt.plot(x1, intercept + slope * x1,'r-')

plt.show
```

Confidence Intervals

```python
from statsmodels.sandbox.regression.predstd import wls_prediction_std

prstd, iv_l, iv_u = wls_prediction_std(results)
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(x,y,'o',label="data")
ax.plot(x,results.fittedvalues,'r-',label="OLS")
ax.plot(x, iv_u, 'r-')
ax.plot(x, iv_l, 'r-')
ax.legend(loc='best')

```

## Nonlinear Fitting

Let's generate points with an arbitrary nonlinear function

$y = \frac{1}{2} x + \frac{1}{2} \sin{x} - \frac{1}{50} (x-5)^2 + 5$

```python
# start with some fake data
nsample = 50
sig = 0.5
# generate linear space
x = np.linspace(0, 20, nsample)
#invent the function
X=np.c_[x, np.sin(x), (x-5)**2, np.ones(nsample)]
# invent co-effecients
beta=[0.5,0.5,-0.02,5.]
#generate
y_true= np.dot(X, beta)
#add noise
y = y_true + sig
```

quick look at the data

```python
plot(x, y_true)
scatter(x,y, c='g')
show()
```

I can access the results attributes

```python
res=sm.OLS(y,X).fit()
```

Let's look at the fit!

## Overfitting

```python
x = np.array([-0.99768, -0.69574, -0.40373, -0.10236, 0.22024, 0.47742, 0.82229])
y = np.array([2.0885, 1.1646, 0.3287, 0.46013, 0.44808, 0.10013, -0.32952])

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x,y, 'o', markersize=12)
plt.xlim((-1.3, 1.3))
```

```python
X = np.c_[x, np.ones(len(x))] # Another way of doing the same thing.
#x = np.vander(x, 1+1)
```

```python
res = sm.OLS(y,X).fit()
res.summary()
```

### Linear fit

```python
plt.figure(figsize=(6,6));
plt.plot(x,y,'o');
xx=np.linspace(-1.25, 1.25, 100)
plt.plot(xx, res.predict(np.vander(xx,3)), 'r-');
plt.title('blie:true, red: OLS')
```

### Quadratic

```python
X = np.c_[x**2, x, np.ones(len(x))]
res = sm.OLS(y, X).fit()
plt.figure(figsize=(6,6));
plt.plot(x,y,'o');
xx=np.linspace(-1.25, 1.25, 100)
plt.plot(xx, res.predict(np.vander(xx,3)), 'r-');
plt.title('blie:true, red: OLS')
```

### Quintic

```python
X = np.c_[x**5, x**4, x**3, x**2, x, np.ones(len(x))]
res = sm.OLS(y, X).fit()

plt.figure(figsize=(6,6));
plt.plot(x,y,'o');
xx=np.linspace(-1.25, 1.25, 100)
plt.plot(xx, res.predict(np.vander(xx,3)), 'r-');
plt.title('blue:true, red: OLS')
```

## Regularization

```python
# Penalizes the curves. # Ridge and LAsso are the different norms for the weights of the curve.

from sklearn.linear_model import Ridge, Lasso
```

```python
ridge = Ridge(alpha=1) # alpha is optional.
ridge.fit(np.vander(x,6), y) # Vander (x,3) , 3 columns and the 1 , WITH THE SQAURED values.

lasso = Lasso(alpha = 0.1)
lasso.fit(np.vander(x,6), y)

plt.figure(figsize=(6,6));
plt.plot(x,y,'o');
xx=np.linspace(-1.25, 1.1, 100) # Linear space we define
plt.plot(xx, res.predict(np.vander(xx,6)), 'r-');
plt.plot(xx, ridge.predict(np.vander(xx,6)), 'g-', label='Ridge');
plt.plot(xx, lasso.predict(np.vander(xx,6)), 'c-', label='Lasso');
plt.legend(loc='best')
```
